## Part 2: Building the Narrative Arc (10 minutes)

Every good data story follows a narrative structure:

```
┌─────────────────────────────────────────────────────────────┐
│                                                             │
│   SITUATION          COMPLICATION         RESOLUTION        │
│   "Here's where      "But there's a       "Here's what      │
│    we are..."         problem..."          we should do..." │
│                                                             │
│        ─────────>  ──────────────>  ──────────────>         │
│                                                             │
└─────────────────────────────────────────────────────────────┘
```

## Part 3: Executive Summary (10 minutes)

An executive summary is the **first thing** decision-makers read. It must:
- State the key finding upfront
- Quantify the impact
- Recommend specific actions
- Fit in one paragraph (or at most two)

### Executive Summary Template

```
[KEY FINDING]: Our analysis of [SCOPE] revealed that [MAIN INSIGHT].

[IMPACT]: This represents [QUANTIFIED IMPACT] to the business.

[ROOT CAUSE]: The primary drivers are [CAUSES], which account for [%] of the issue.

[RECOMMENDATION]: We recommend [SPECIFIC ACTIONS] which could [EXPECTED OUTCOME].
```

## Key Takeaways

1. **Lead with insight, not data** - Start with what matters, then support with evidence
2. **Use the narrative arc** - Situation → Complication → Resolution guides your audience
3. **Quantify impact** - Numbers make your story credible and urgent
4. **Be actionable** - Always end with what the audience should DO
