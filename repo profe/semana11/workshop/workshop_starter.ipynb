## Part 1: Story Structure Templates

Use these templates to organize your thoughts before creating slides.

### Template A: The Problem-Solution Story

Best for: Business analytics, process improvement, optimization projects

```
HOOK: "[Stakeholder] is losing [specific amount] because of [problem]."

SITUATION: 
- We analyzed [data source] covering [time period / scope]
- The goal was to understand [business question]
- Current state: [baseline metrics]

KEY INSIGHT:
- [Single most important finding]
- This explains [X]% of [the problem]
- Visualization: [describe the chart that proves this]

EVIDENCE:
- Supporting finding 1: [detail]
- Supporting finding 2: [detail]
- We ruled out [alternative explanation] because [reason]

CALL TO ACTION:
- Recommendation 1: [specific action] → Expected impact: [metric]
- Recommendation 2: [specific action] → Expected impact: [metric]
- Next step: [immediate action to take]
```

### Template B: The Discovery Story

Best for: Exploratory analysis, research projects, pattern discovery

```
HOOK: "We discovered something unexpected in [data source]."

SITUATION:
- We started by exploring [data source]
- Initial hypothesis: [what we expected to find]
- Method: [brief description of approach]

KEY INSIGHT:
- What we actually found: [surprising discovery]
- This challenges the assumption that [conventional wisdom]
- Visualization: [describe the chart that reveals this]

EVIDENCE:
- The pattern holds across [different segments/conditions]
- Statistical significance: [confidence level/p-value if applicable]
- We validated by [cross-check method]

CALL TO ACTION:
- This means we should reconsider [current practice]
- Opportunity: [what this enables]
- Further research needed: [open questions]
```

### Template C: The Comparison Story

Best for: A/B testing, benchmarking, competitive analysis

```
HOOK: "[Option A] outperforms [Option B] by [X]%, and here's why."

SITUATION:
- We compared [A] vs [B] using [metrics]
- Data covered [scope and time period]
- Both groups had [similar baseline characteristics]

KEY INSIGHT:
- Winner: [A or B] with [specific advantage]
- The difference is [statistically significant / practically meaningful]
- Visualization: [side-by-side comparison chart]

EVIDENCE:
- Breakdown by segment: [where differences are largest]
- Controlled for: [confounding variables]
- Edge cases: [where the pattern doesn't hold]

CALL TO ACTION:
- Adopt [winner] for [specific use case]
- Expected improvement: [projected metrics]
- Rollout plan: [phases/timeline]
```

### Visualization Selection Guide

Choose the right chart for your message:

| If you want to show... | Use... | Example |
|------------------------|--------|----------|
| Comparison between categories | Bar chart | Sales by region |
| Change over time | Line chart | Monthly revenue trend |
| Part-to-whole relationship | Pie/Donut chart (sparingly) | Market share |
| Distribution | Histogram or box plot | Age distribution |
| Correlation | Scatter plot | Price vs. demand |
| Geographic patterns | Map | Sales by state |
| Process or flow | Sankey or flowchart | Customer journey |
| Ranking | Horizontal bar chart | Top 10 products |

## Appendix: Quick Reference

### The 5-Slide Structure

```
┌──────────────────────────────────────────────────────────────────────┐
│                                                                      │
│  1. HOOK          2. SITUATION     3. KEY INSIGHT                    │
│  "Grab attention" "Set context"    "Main finding"                    │
│  [30 sec]         [45 sec]         [90 sec]                          │
│                                                                      │
│                   4. EVIDENCE       5. CALL TO ACTION                │
│                   "Prove it"        "What to do"                     │
│                   [90 sec]          [30 sec]                         │
│                                                                      │
└──────────────────────────────────────────────────────────────────────┘
```

### Storytelling Checklist

- [ ] My hook is memorable and relevant
- [ ] I explain WHY this matters, not just WHAT I found
- [ ] I have ONE clear key insight (not a list)
- [ ] My visualizations are simple and readable
- [ ] I end with specific, actionable recommendations
- [ ] My pitch fits in 5 minutes

### Common Mistakes to Avoid

1. Starting with methodology instead of the problem
2. Including too many findings (pick your best one)
3. Charts with too much detail or unclear labels
4. Ending with "any questions?" instead of a strong close
5. Reading from slides instead of telling a story
